```python
import pandas as pd
import cond as cd
import DP_Sliding as dp
import process_edited as pce
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Length of the time series
n = 10000

# Generate time points
time_points = np.arange(n)

# Create a sinusoidal temperature pattern with noise
np.random.seed(42)  # For reproducibility
seasonal_pattern = 10 * np.sin(2 * np.pi * time_points / 365)  # Yearly cycle
noise = np.random.normal(0, 2, n)  # Random noise
temperature = 15 + seasonal_pattern + noise  # Base temp of 15Â°C

# Generate WeatherCondition based on Temperature rules
weather_condition = np.where(
    temperature > 20, "Sunny",
    np.where(temperature > 10, "Cloudy",
             np.where(temperature > 0, "Rainy", "Snowy")))

# Create hourly timestamps starting from '10/2/2012 9:00'
start_time = pd.Timestamp("2012-10-02 09:00")
timestamps = pd.date_range(start=start_time, periods=n, freq='H')

# Create a DataFrame for the time series
real_df = pd.DataFrame({
    "date": timestamps,
    "Temp": temperature,
    "Weather": weather_condition
})
```

```python
real_df1 = real_df.drop(['date'], axis=1)
parser = pce.DataFrameParser().fit(real_df1, 1)
name = parser.column_name()
```

```python
response_list = ['Temp']; VAE_training = 5000; diff_training = 10000;

processed_data, time_info, train_idx, test_idx = dp.single_split_train_test(real_df, 48)
response_df, con_df, response_train, response_test, cond_train, cond_test, time_info_train, time_info_test = dp.divide_respons_cond_time(real_df, processed_data, time_info, response_list, train_idx, test_idx)
model = cd.C_TimeAutoDiff(response_df, con_df, response_train, cond_train, time_info_train, VAE_training, diff_training)
```
